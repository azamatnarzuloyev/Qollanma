```python
x = np.arange(0, 4 * np.pi, 0.1)
sin = np.sin(x) # sin(x)
cos = np.cos(x) # cos(x)

plt.figure(figsize=(12,6))
# sin(x) grafigi
plt.plot(x, sin,
         linewidth=1.5,
         linestyle='-.',
         color='#8E44AD',
         alpha=0.7,
         marker='h')

# cos(x) grafigi
plt.plot(x, cos,
         linewidth=1.5,
         linestyle=':',
         color='c',
         alpha=0.7,
         marker='v')

# Grafik nomi:
plt.title('sin(x) va cos(x) funksiya grafigi')
# x-o'q nomi:
plt.xlabel("pi")
# Har bir chiziq haqida ma'lumot:
plt.legend(['sin(x)','cos(x)'], loc=0, fontsize='x-large')
plt.grid(color='tomato', alpha=0.5, linestyle='--') # setkani o'zgartirish
plt.show()
```

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_html('https://en.tutiempo.net/climate/ws-384570.html')
```

```python
#@title To'g'ri javob
tables = pd.read_html("https://en.tutiempo.net/climate/ws-384570.html", index_col=0)
# read_html sahifadagi har bir jadvallarni bitta ro'yxatga joylaydi, 
# ro'yxatni tahlil qilib bizga kerak jadval [3] indeksda joylashganini bilsak bo'ladi
df = tables[3][tables[3].index>1990]
# df.head()
```

```python
df.head()
```

```python
df.to_csv('data.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.replace("-", np.nan)
```

```python
df = df.astype(dtype='float')
```

```python
df.head()
```

```python
df.isnull().sum()/len(df)*100
```

```python
df.sample(10)
```

```python
df = df.drop(columns='PP')
```

```python
df.loc[2000:2005]
```

```python
df.fillna(method='ffill', inplace=True)
```

```python
df.to_csv('data.csv')
```

```python
from matplotlib import markers
#@title To'g'ri javob
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(16,6))
plt.title("Toshkent shahar uchun o'rtacha harorat")
plt.xlabel('yil')
plt.ylabel('harorat')
x= df['T']
plt.plot(x, linewidth=2, color='blue', marker='h', linestyle='--')
plt.show()
```

```python
minT = df['Tm']
maxT = df['TM']
plt.figure(figsize=(16,8))
plt.plot(minT, linewidth=2, linestyle="--", marker='h', markerfacecolor='g', label="O'rtacha minimum harorat")
plt.plot(maxT, linewidth=2, linestyle="-", marker='v', markerfacecolor='r', label="O'rtacha maksimum harorat")
plt.grid()
plt.legend()
plt.title("Toshkent shahri uchun yillik o'rtacha min va max harorat")
plt.xlabel("Yil")
plt.ylabel("Harorat")
plt.savefig('tempMinMax.png')
plt.show()
```
